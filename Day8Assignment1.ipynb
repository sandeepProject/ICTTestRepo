Case Study 1: Employee Payroll System
Problem Statement:
Design a simple Payroll Management System using OOP concepts to manage employees and their salaries efficiently. The system should handle salary calculations for different types of employees.
Key Requirements:
1. Class Definitions: • Employee: Base class with attributes like name, emp_id, and base_salary. • FullTimeEmployee and PartTimeEmployee: Derived classes that calculate pay differently.
2. Inheritance & Polymorphism: • Implement a method calculate_pay() that works differently for full-time and part-time employees.
3. Encapsulation: • Keep the salary details private and provide getter methods.
4. __repr__ Usage: • Implement a __repr__() method for clean output of employee details.

```python
# Base Class
class Employee:
  def __init__(self,name,emp_id,base_salary):
    self.name=name
    self.emp_id=emp_id
    self.__base_salary=base_salary # __base_salary --> private attribute (Encapsulation)
  def get_base_Salary(self):       # getter method
    return self.__base_salary
  def __repr__(self):              # __repr__ usage
    return f"Employee(Name: {self.name}, ID: {self.emp_id}, Base Salary: ₹{self.__base_salary:.2f})"
```

```python
#emp=Employee("Hari","EMP101",50000)
#repr(emp)
```

```python
# Derived Class : 1 FullTimeEmployee
class FullTimeEmployee(Employee):
# full-time employee's pay based on base salary.
  def calculate_pay(self):            # method overriding calculate_pay --> polymorphism
    return self.get_base_Salary()
  def __repr__(self):
    details=super().__repr__()
    return f"Full Time Employee {details.replace('Employee','')}"
```

```python
# Derived Class : 2 PartTimeEmployee
class PartTimeEmployee(Employee):
  def __init__(self,name,emp_id,base_salary,hours_worked,hourly_rate):
    super().__init__(name,emp_id,base_salary)
    self.hours_worked=hours_worked
    self.hourly_rate=hourly_rate

# part-time employee's pay based on hours worked.
  def calculate_pay(self):                  # method overriding calculate_pay --> polymorphism
    partime_Salary =self.hours_worked * self.hourly_rate
    return partime_Salary

  def __repr__(self):             # __repr__() usage
    details=super().__repr__()
    return f"Part Time Employee(Name: {self.name}, ID: {self.emp_id}, Salary: ₹{self.calculate_pay()}, Hours: {self.hours_worked}, Rate: ₹{self.hourly_rate:.2f})"
```

```python
# Details of different employee
employee1 = FullTimeEmployee(name="Megha", emp_id="ICT101", base_salary=50000)
employee2 = PartTimeEmployee(name="Aravind", emp_id="ICT202", base_salary=0, hours_worked=80, hourly_rate=25)
employee3 = FullTimeEmployee(name="Irne", emp_id="ICT103", base_salary=55000)
employee4 = PartTimeEmployee(name="Sandeep", emp_id="ICT205", base_salary=0, hours_worked=50, hourly_rate=25)
# Employees to a list
employees = [employee1, employee2, employee3,employee4]
```

```python
print("--- Payroll System Report ---")
for emp in employees:
    # repr() usage provides clean output of employee details
    print(f"\nDetails: {repr(emp)}")

    # Polymorphism: call calculate_pay()
    pay = emp.calculate_pay()
    print(f"Calculated Pay: ₹{pay:.2f}")
```
