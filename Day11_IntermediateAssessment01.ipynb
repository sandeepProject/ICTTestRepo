2.1 Calculate the total revenue for the store (sum of all quantities ×
price per item).


```python
# Calculating Total Revenue for the store
total_Revenue = 0.0
for sale in sales_Data:
    order_Total = sale['quantity'] * sale['price_per_item']
    total_Revenue += order_Total

print(f"Total Revenue: {total_Revenue}")
```

2.2 Identify all orders where the total order value (quantity × price
per item) exceeds ₹2,000. Store the result as a new list.


```python
exceeded_value_Orders = []
base_Rate = 2000.0

for sale in sales_Data:
    order_Value = sale['quantity'] * sale['price_per_item']
    if order_Value > base_Rate:
        # New dictionary containing order details of cutomers that exceeded base rate > 2,000.
        exceeded_value_order_List = {
            'order_id': sale['order_id'],
            'customer': sale['customer'],
            'item': sale['item'],
            'quantity': sale['quantity'],
            'price_per_item': sale['price_per_item'],
            'order_date': sale['order_date'],
            'order_value': order_Value
        }
        exceeded_value_Orders.append(exceeded_value_order_List)

print("High-Value Orders (total value > ₹2,000):")
for order in exceeded_value_Orders:
    print(order)
```

2.3 Find the item with the highest total quantity sold across all
orders.


```python
# Step 1: New dictionary for quantities
item_quantities_Dict = {}

for sale in sales_Data:
    item = sale['item']
    quantity = sale['quantity']

    item_quantities_Dict[item] = item_quantities_Dict.get(item, 0) + quantity

# Hint.
# To find the item with the maximum quantity
# Laptop: Sold 3 times with 1 unit each = 3 units.
# Smartphone: Sold 2 times with 1 and 2 units = 3 units.
# Headphones: Sold 2 times with 2 and 1 units = 3 units.
# Charger: Sold 1 time with 3 units.

#print(item_quantities_Dict)
best_selling_item = max(item_quantities_Dict, key=item_quantities_Dict.get)
#print(type(best_selling_item))
#print(item_quantities_Dict[best_selling_item])
total_Qty = item_quantities_Dict[best_selling_item]
#print(total_Qty)
print(f"The Best Selling Item is '{best_selling_item}' with {total_Qty} units sold.")
```

3.1 Write a function apply_discount(orders, discount_rate) that takes
a list of orders and applies a percentage discount (e.g., 10% = 0.10) to
the price_per_item for each order.
Return a new list of orders (do not modify the original list).

```python
import copy    # to copy the orginal data sale_Date to discount_orders
def apply_discount(orders, discount_rate):
    # deep copy to ensure the original data remains unchanged
    discounted_Orders = copy.deepcopy(orders)

    for order in discounted_Orders:
        original_Price = order['price_per_item']
        discount_Amount = original_Price * discount_rate
        new_Price = original_Price - discount_Amount
        order['price_per_item'] = round(new_Price, 2) # Update the price_per_item in the new list of orders.
    return discounted_Orders
```

```python
discount_Rate=0.10
discounted_sales_data = apply_discount(sales_Data, discount_Rate) # calling created function.
#print(discounted_sales_data)
#print(sales_Data)
print("Original Data sale_Data:")
print(sales_Data[0])
print("\nDiscounted Data with 10% discount :")
print(discounted_sales_data[0])
```

3.2 Explain, with a code example in your script, the difference
between local and global variable scope in Python.

```python
# --- GLOBAL SCOPE ---
# This variable is accessible throughout the entire script
store_name = "ShopEase"

def global_Local_Variable(order_list):
    global store_name     #global variable declaration
    store_name = "ShopEase India"  # we can also change the value inside the fuction
    # --- LOCAL SCOPE ---
    # This variable (discount_status) accessible only inside this function or scope
    order_count = len(order_list)
    discount_status = "Applied"

    #print(f"[{store_name}] : Processing {order_count} orders. Discount Stataus  : {discount_status}")
    return f"Details : {store_name}: {order_count} orders processed; dicount status : {discount_status}"
```

```python
# Checking the scopes
print(global_Local_Variable(sales_Data))

# Accessing the Global variable :
print(f"Welcome to the {store_name}")

# Accessing the Local variable will throw error:
try:
    print(discount_status)
except NameError as e:  # to catch exception
    print(f"Error: {e} (This is because 'discount_status' is local to the function)")
```

4.1 Write a custom iterator class OrderIterator to iterate through the
orders, yielding one order at a time.

```python
class OrderIterator:
    def __init__(self, orders): # Initialisation function
        # Store the list of orders and set an index = 0
        self._orders = orders
        self._index = 0
    def __iter__(self): # Iterate function ->This makes the class an iterable, allowing it to be used directly in for loops.
        return self
    def __next__(self): # It fetches the current item and increments the pointer. If the pointer exceeds the length of the list, it raises StopIteration
      # try:
        if self._index < len(self._orders):
            order = self._orders[self._index]
            self._index += 1
            return order
        else:
            # Raise StopIteration to signal the end of the loop
            raise StopIteration
      # except IndexError:
      #   raise StopIteration
```

```python
order_iter = OrderIterator(sales_Data)

print("Iterating through orders using OrderIterator:")
for order in order_iter:
    print(f"Order ID: {order['order_id']} | Customer: {order['customer']}")
```

4.2 Create a dictionary grouping all orders by customer using a
dictionary comprehension

```python
# Customer names
customers = {order['customer'] for order in sales_Data}
#print(customers)
# Dictionary comprehension to group the orders
 #   {key_expression: value_expression for item in iterable [if condition]}
grouped_orders = {
    name: [order for order in sales_Data if order['customer'] == name] for name in customers
}
#print(grouped_orders)

for customer, orders in grouped_orders.items():
    print(f"{customer}: {len(orders)} order's")
```
