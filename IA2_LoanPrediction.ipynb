```python
# Preprocessing and Feature Eng.
def preprocess_data(df):
    df = df.copy()
    df['Dependents'] = df['Dependents'].replace('3+', 3).astype(float)
    binary_map = {'Male': 1, 'Female': 0, 'Yes': 1, 'No': 0, 'Graduate': 1, 'Not Graduate': 0}
    for col in ['Gender', 'Married', 'Education', 'Self_Employed']:
        df[col] = df[col].map(binary_map)
    df['Property_Area'] = df['Property_Area'].map({'Urban': 2, 'Semiurban': 1, 'Rural': 0})
    df['Total_Income'] = df['ApplicantIncome'] + df['CoapplicantIncome']
    #df['Total_Income_Log'] = np.log(df['Total_Income'])
    df = df.drop(['ApplicantIncome', 'CoapplicantIncome'], axis=1)

    return df.drop('Loan_ID', axis=1, errors='ignore')
```

```python
# Prepare X and y
X = preprocess_data(train_df).drop('Loan_Status', axis=1)
y = train_df['Loan_Status'].map({'Y': 1, 'N': 0})
X_test_final = preprocess_data(test_df)
```

```python
X.isna().sum()
```

```python
# KNN IMPUTATION
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5)
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)
X_test_imputed = pd.DataFrame(imputer.transform(X_test_final), columns=X.columns)

# Round categorical results
round_cols = ['Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area', 'Dependents', 'Credit_History']
for col in round_cols:
    X_imputed[col] = X_imputed[col].round()
    X_test_imputed[col] = X_test_imputed[col].round()
```

```python
# 3. DEFINE MODELS
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, solver='liblinear'),
    "SVM (Radial)": SVC(kernel='rbf', probability=True),
    "Random Forest": RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42),
    "XGBoost": XGBClassifier(eval_metric='logloss', learning_rate=0.1, max_depth=4),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3)
}
```

```python
# EVALUATE ALL MODELS
results = {}
print("--- Model Accuracy Comparison (Cross-Validation) ---")
for name, model in models.items():
    # Used 5-fold cross-validation to get a stable accuracy score
    cv_scores = cross_val_score(model, X_imputed, y, cv=5, scoring='accuracy')
    results[name] = cv_scores.mean()
    print(f"{name}: {cv_scores.mean():.4f}")
```

```python
# SELECT BEST MODEL & GENERATE SUBMISSION
best_model_name = max(results, key=results.get)
print(f"\nBest Model: {best_model_name}")

best_model = models[best_model_name]
best_model.fit(X_imputed, y)
final_preds = best_model.predict(X_test_imputed)
```

```python
# EXPORT
submission = pd.DataFrame({
    'Loan_ID': test_ids,
    'Loan_Status': ['Y' if p == 1 else 'N' for p in final_preds]
})
submission.to_csv('CV_optimized_submission.csv', index=False)
print("SUCCESS: 'CV_optimized_submission.csv' generated.")
```

```python
# DEFINE MODELS AND PARAMETER GRIDS
from sklearn.model_selection import GridSearchCV
model_params = {
    "Logistic_Regression": {
        "model": LogisticRegression(max_iter=1000, solver='liblinear'),
        "params": {
            "C": [0.01, 0.1, 1, 10],
            "penalty": ['l1', 'l2']
        }
    },
    "Random_Forest": {
        "model": RandomForestClassifier(random_state=42),
        "params": {
            "n_estimators": [50, 100, 200],
            "max_depth": [3, 5, 7],
            "criterion": ['gini', 'entropy']
        }
    },
    "XGBoost": {
        "model": XGBClassifier(eval_metric='logloss', random_state=42),
        "params": {
            "n_estimators": [50, 100],
            "max_depth": [3, 4, 5],
            "learning_rate": [0.01, 0.1, 0.2]
        }
    }
}

# RUN GRID SEARCH FOR EACH MODEL
best_models = {}
scores = []

print("--- Starting Hyperparameter Tuning (GridSearchCV) ---")
for model_name, mp in model_params.items():
    grid = GridSearchCV(mp['model'], mp['params'], cv=5, scoring='accuracy', n_jobs=-1)
    grid.fit(X_imputed, y)

    best_models[model_name] = grid.best_estimator_
    scores.append({
        'model': model_name,
        'best_score': grid.best_score_,
        'best_params': grid.best_params_
    })
    print(f"Finished {model_name}. Best Score: {grid.best_score_:.4f}")

# SELECT WINNER AND GENERATE SUBMISSION
results_df = pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
top_model_name = results_df.sort_values(by='best_score', ascending=False).iloc[0]['model']
print(f"\nOverall Winning Model: {top_model_name}")

# Generate predictions using the best tuned model
winner_model = best_models[top_model_name]
final_preds_GSCV = winner_model.predict(X_test_imputed)
```

```python
# EXPORT
submission = pd.DataFrame({
    'Loan_ID': test_ids,
    'Loan_Status': ['Y' if p == 1 else 'N' for p in final_preds_GSCV]
})
submission.to_csv('GSCV_optimized_submission.csv', index=False)
print("SUCCESS: 'GSCV_optimized_submission.csv' generated.")
```
