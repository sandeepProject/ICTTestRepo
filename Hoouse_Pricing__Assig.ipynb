```python
# Sales Price - Droping the rows contains null values because it is the target column

df.dropna(subset=['Sale Price'], inplace=True)

print(f"Missing Sale Price values: {df['Sale Price'].isnull().sum()}")
```

```python
from sklearn.impute import SimpleImputer
from sklearn.impute import KNNImputer
```

```python
# Define the numerical columns with missing values
num_cols = [
    'No of Bathrooms', 'Flat Area (in Sqft)', 'Lot Area (in Sqft)',
    'Area of the House from Basement (in Sqft)', 'Living Area after Renovation (in Sqft)'
]

# Initialize the Imputer with Median strategy
imputer = SimpleImputer(strategy='median')

# Fit and transform the columns
df[num_cols] = imputer.fit_transform(df[num_cols])

# Verify
print(df[num_cols].isnull().sum())
```

```python
# Select the target columns and a few predictor columns for better accuracy
knn_cols = ['Zipcode', 'Latitude', 'Longitude', 'Flat Area (in Sqft)', 'Sale Price']

# Initialize the KNNImputer imputer
# n_neighbors=5 is standard, weights='distance' gives closer neighbors more influence
imputer = KNNImputer(n_neighbors=5, weights='distance')

# Fit and transform the columns
df[knn_cols] = imputer.fit_transform(df[knn_cols])

# round() - Zipcode back to an integer
df['Zipcode'] = df['Zipcode'].round().astype(int)

print(df[['Zipcode', 'Latitude', 'Longitude']].isnull().sum())
```

```python
# List of columns to log transform
feature_cols = ['Sale Price', 'Flat Area (in Sqft)', 'Lot Area (in Sqft)']
```

```python
df[feature_cols].hist(bins=50, figsize=(10, 5))
```

```python
# Apply the log transformation
for col in feature_cols:
    # Use np.log1p for sale price (log(1+x)) as a safe practice to handle potential zero values
    if col == 'Sale Price':
        df[col] = np.log1p(df[col])
    else:
        # Simple log for areas
        df[col] = np.log(df[col])
```
