1. Basic Array Operations Convert the mpg column into a NumPy array and calculate:<br>
○ The mean, median, and standard deviation of mpg.<br>
○ The number of cars with mpg greater than 25.

```python
# importing numpy and pandas
import numpy as np
import pandas as pd
```

```python
# reading data from csv file
data=pd.read_csv('auto-mpg.csv')
data.head(4)

# convert the mpg column into numpy array
mpg_numpy_array=data['mpg'].to_numpy()

# calculate mean, median and standard deviation
data_mean = np.mean(mpg_numpy_array)
data_median= np.median(mpg_numpy_array)
data_std = np.std(mpg_numpy_array)
print(f'Mean : {data_mean:.2f} , Median : {data_median:.2f} , Standard Deviation : {data_std:.2f}')

# No: of cars with mpg >25
count =np.sum(mpg_numpy_array>25)
print(f'No: of cars with mpg >25 : {count}')
```

10. Finding Anomalies<br> Identify potential outliers in the mpg column using the Interquartile Range (IQR) method. Specifi cally:<br>
● Calculate the IQR for mpg.<br>
● Defi ne outliers as values less than Q1 - 1.5 * IQR or greater than Q3 + 1.5 * IQR.<br>
● Create a DataFrame of cars classifi ed as outliers, displaying car_name, mpg, and model_year.

```python
#df['mpg'].describe()
Q1 = df['mpg'].quantile(0.25)
Q3 = df['mpg'].quantile(0.75)

# Interquartile Range (IQR)
IQR = Q3 - Q1

print(f"Q1 (25%): {Q1}")
print(f"Q3 (75%): {Q3}")
print(f"IQR: {IQR}")
```

```python
lower_values = Q1 - 1.5 * IQR
upper_values = Q3 + 1.5 * IQR
outliers = df[(df['mpg'] < lower_values) | (df['mpg'] > upper_values)]

print(f"Lower Values: {lower_values}")
print(f"Upper Values: {upper_values}")
print(f"Number of outliers found: {len(outliers)}")
print(outliers[['car name', 'mpg','model year']])
```
